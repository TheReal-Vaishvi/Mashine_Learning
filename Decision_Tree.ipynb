```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
data = {
    'Weather':['Sunny','Cloudy','Sunny','Cloudy','Rainy',
               'Rainy','Rainy','Sunny','Cluody','Rainy'],
    'Temperature':['Hot','Hot','Mild','Mild','Mild',
                   'Cool', 'Mild','Hot','Hot','Mild'],
    'Humidity' : ['High','High','Normal','High','High',
                  'Normal','High','High','Normal','High'],
    'Wind' : ['Weak','Weak','Strong','Strong','Strong',
              'Strong','Weak','Strong','Weak','Strong'],
    'Play' : ['No','Yes','Yes','Yes','No',
              'No','Yes','No','Yes','No']
}
```

```python
df = pd.DataFrame(data)

df
```

```python
#Encode Categorical Values
encoder = {}

for col in df.columns:
  le = LabelEncoder()
  df[col] = le.fit_transform(df[col])
  encoder[col] = le

df
```

```python
x = df.drop('Play', axis=1)

y = df['Play']
```

```python
print("Features")
print(x)
```

```python
model = DecisionTreeClassifier(
    criterion='entropy',
    random_state=42
)
```

```python
#Train Model

model.fit(x,y)

print("Model Trained Successfully")
```

```python
#Dispay information gain importance

importance = model.feature_importances_

for feature, score in zip(x.columns, importance):
  print(feature, ":" , score)
```

```python
#Predict New Data
new_data = pd.DataFrame({
    'Weather' : ['Sunny'],
    'Temperature' : ['Mild'],
    'Humidity' : ['Normal'],
    'Wind' : ['Strong']
})
```

```python
new_data
```

```python
for col in new_data.columns:
  new_data[col] = encoder[col].transform(new_data[col])

predication = model.predict(new_data)
result = encoder['Play'].inverse_transform(predication)

print("Result : ", result[0])
```

```python
plt.figure(figsize=(8,6))

plot_tree(
    model,
    feature_names=x.columns,
    class_names=['No','Yes'],
    filled=True,
    rounded=True,
)

plt.title("Decision Tree - Play Tennis Dataset")

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.bar(
    x.columns,
    model.feature_importances_
)

plt.title("Feature Importance - Play Tennis Dataset")
plt.xlabel("Features")
plt.ylabel("Importance")

plt.show()
```

```python
#Accuracy

from sklearn.metrics import accuracy_score

y_pred = model.predict(x)

accuracy = accuracy_score(y, y_pred)

print("Accuracy : ", accuracy*100, "%")
```
